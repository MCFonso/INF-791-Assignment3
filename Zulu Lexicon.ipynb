```python
# Let's use a translation service to get you started
import requests
import time
import pandas as pd

def translate_word(word, source_lang='fr', target_lang='zu'):
    """Translate a single word using MyMemory API"""
    try:
        url = "https://api.mymemory.translated.net/get"
        params = {
            'q': word,
            'langpair': f"{source_lang}|{target_lang}"
        }
        response = requests.get(url, params=params, timeout=10)
        if response.status_code == 200:
            data = response.json()
            if data['responseStatus'] == 200:
                return data['responseData']['translatedText']
        return f"[Need manual translation: {word}]"
    except:
        return f"[Translation error: {word}]"

print("Starting automatic translation of first 20 words...")

# Create a copy to work with
translated_batch = test_batch.copy()

# Translate first 20 words
for i in range(20):
    french_word = translated_batch.iloc[i]['French']
    zulu_translation = translate_word(french_word, 'fr', 'zu')
    translated_batch.iloc[i, translated_batch.columns.get_loc('Zulu')] = zulu_translation
    print(f"{i+1}. {french_word}  {zulu_translation}")
    time.sleep(1)  # Be respectful to the API

print("\nFirst 20 words translated!")
display(translated_batch.head(20))

# Save the partially translated batch
translated_batch.to_excel('partially_translated_test_batch.xlsx', index=False)
print("\nSaved 'partially_translated_test_batch.xlsx'")
```

```python
# Create a comprehensive progress tracking system
def create_progress_tracker(total_words, completed_count=0):
    """Create a visual progress tracker"""
    progress_percent = (completed_count / total_words) * 100
    
    print("=== TRANSLATION PROGRESS TRACKER ===")
    print(f"Overall Progress: {completed_count}/{total_words} words ({progress_percent:.1f}%)")
    
    # Visual progress bar
    bar_length = 50
    filled_length = int(bar_length * completed_count // total_words)
    bar = '' * filled_length + '' * (bar_length - filled_length)
    print(f"[{bar}] {progress_percent:.1f}%")
    
    # Milestones
    milestones = [25, 50, 75, 100]
    for milestone in milestones:
        words_needed = int(total_words * milestone / 100)
        if completed_count >= words_needed:
            status = " COMPLETED"
        else:
            remaining = words_needed - completed_count
            status = f" {remaining} words to go"
        print(f" {milestone}% ({words_needed} words): {status}")
    
    # Time estimation
    remaining_words = total_words - completed_count
    estimated_hours = remaining_words / 50  # Assuming 50 words per hour
    print(f"\n Estimated time remaining: {estimated_hours:.1f} hours")
    
    return {
        'completed': completed_count,
        'total': total_words,
        'percentage': progress_percent,
        'remaining_hours': estimated_hours
    }

# Update progress (we've done 20 words so far)
progress = create_progress_tracker(total_words, 20)
```

```python
print(" CREATING COMPREHENSIVE BATCH PROCESSING SYSTEM FOR ALL 6963 WORDS")

import pandas as pd
import time
import requests
import os
from tqdm.notebook import tqdm  # For progress bars

# Initialize tqdm for pandas
tqdm.pandas()

def robust_translate_word(word, source_lang='fr', target_lang='zu', max_retries=3):
    """
    Robust translation function with retries and multiple fallback methods
    """
    if pd.isna(word) or word == '':
        return ''
    
    word = str(word).strip()
    
    # Skip very long texts
    if len(word) > 50:
        return f"[Too long: {word[:20]}...]"
    
    for attempt in range(max_retries):
        try:
            # Method 1: MyMemory API
            url = "https://api.mymemory.translated.net/get"
            params = {
                'q': word,
                'langpair': f"{source_lang}|{target_lang}"
            }
            response = requests.get(url, params=params, timeout=15)
            
            if response.status_code == 200:
                data = response.json()
                if data['responseStatus'] == 200:
                    translation = data['responseData']['translatedText']
                    # Basic validation
                    if translation and translation != word and len(translation) > 0:
                        return translation
            
            # If we get here, translation failed
            time.sleep(2)  # Wait before retry
                
        except Exception as e:
            if attempt < max_retries - 1:
                time.sleep(2)  # Wait before retry
                continue
            else:
                return f"[Error: {word}]"
    
    return f"[Failed: {word}]"

def process_batch(batch_df, batch_number, total_batches, start_index):
    """
    Process a single batch with progress tracking
    """
    print(f" Processing Batch {batch_number}/{total_batches} (words {start_index + 1}-{start_index + len(batch_df)})")
    
    # Create a copy to avoid modifying original
    processed_batch = batch_df.copy()
    
    # Translate each word in the batch
    translations = []
    total_words = len(processed_batch)
    
    for i, (index, row) in enumerate(processed_batch.iterrows()):
        french_word = row['French']
        
        # Show progress every 25 words
        if (i + 1) % 25 == 0:
            print(f"   Progress: {i + 1}/{total_words} words")
        
        # Translate the word
        translation = robust_translate_word(french_word, 'fr', 'zu')
        translations.append(translation)
        
        # Respectful delay - increased to avoid rate limiting
        time.sleep(2.5)
    
    # Add translations to the batch
    processed_batch['Zulu'] = translations
    
    return processed_batch

# Load the complete dataset
try:
    df = pd.read_excel('lexicon_6000_words.xlsx')
    print(f" Loaded dataset with {len(df)} words")
except Exception as e:
    print(f" Error loading file: {e}")
    # If file doesn't exist, we need to handle this
    raise e

# Add Zulu column if it doesn't exist
if 'Zulu' not in df.columns:
    df['Zulu'] = ''

total_words = len(df)
batch_size = 50  # Smaller batches to avoid timeouts
num_batches = (total_words + batch_size - 1) // batch_size

print(f" BATCH PROCESSING PLAN:")
print(f"   Total words: {total_words}")
print(f"   Batch size: {batch_size} words")
print(f"   Total batches: {num_batches}")
print(f"   Estimated time: {(total_words * 2.5) / 3600:.1f} hours")
print(f"   Output directory: 'all_batches_complete/'")

# Create output directory
os.makedirs('all_batches_complete', exist_ok=True)
```

```python
print(" STARTING COMPLETE BATCH PROCESSING")

# Create a batch tracking system
batch_tracking = {}

for batch_num in range(num_batches):
    start_idx = batch_num * batch_size
    end_idx = min((batch_num + 1) * batch_size, total_words)
    batch_range = f"{start_idx + 1}-{end_idx}"
    
    output_filename = f'all_batches_complete/batch_{batch_num + 1:03d}_of_{num_batches}_{batch_range}.xlsx'
    status_filename = f'all_batches_complete/batch_{batch_num + 1:03d}_status.txt'
    
    batch_tracking[batch_num + 1] = {
        'range': batch_range,
        'output_file': output_filename,
        'status_file': status_filename,
        'status': 'PENDING'
    }

# Save batch tracking
tracking_df = pd.DataFrame.from_dict(batch_tracking, orient='index')
tracking_df.to_excel('all_batches_complete/batch_tracking.xlsx')
print(" Created batch tracking system")

# Process batches
completed_batches = 0

for batch_num in range(num_batches):
    batch_info = batch_tracking[batch_num + 1]
    output_file = batch_info['output_file']
    
    # Skip if already processed
    if os.path.exists(output_file):
        print(f" Batch {batch_num + 1} already exists, skipping...")
        completed_batches += 1
        continue
    
    # Update status
    batch_tracking[batch_num + 1]['status'] = 'PROCESSING'
    
    try:
        # Extract batch data
        start_idx = batch_num * batch_size
        end_idx = min((batch_num + 1) * batch_size, total_words)
        batch_data = df.iloc[start_idx:end_idx].copy()
        
        # Process the batch
        processed_batch = process_batch(
            batch_data, 
            batch_num + 1, 
            num_batches, 
            start_idx
        )
        
        # Save the batch
        processed_batch.to_excel(output_file, index=False)
        
        # Update status
        batch_tracking[batch_num + 1]['status'] = 'COMPLETED'
        completed_batches += 1
        
        # Save progress
        progress = (completed_batches / num_batches) * 100
        print(f" Batch {batch_num + 1} completed! Overall progress: {progress:.1f}%")
        
        # Save updated tracking
        tracking_df = pd.DataFrame.from_dict(batch_tracking, orient='index')
        tracking_df.to_excel('all_batches_complete/batch_tracking.xlsx')
        
        # Small delay between batches
        time.sleep(5)
        
    except Exception as e:
        print(f" Error processing batch {batch_num + 1}: {e}")
        batch_tracking[batch_num + 1]['status'] = f'ERROR: {str(e)[:50]}'
        
        # Save error status
        tracking_df = pd.DataFrame.from_dict(batch_tracking, orient='index')
        tracking_df.to_excel('all_batches_complete/batch_tracking.xlsx')
        
        # Continue with next batch despite error
        continue

print(f" BATCH PROCESSING COMPLETED: {completed_batches}/{num_batches} batches processed")
```

```python
print(" COMBINING ALL BATCHES INTO MASTER FILE")

def combine_all_batches():
    """Combine all batch files into a single master file"""
    all_batches = []
    
    # Find all batch files
    batch_files = [f for f in os.listdir('all_batches_complete') 
                  if f.startswith('batch_') and f.endswith('.xlsx')]
    batch_files.sort()  # Sort to maintain order
    
    print(f"Found {len(batch_files)} batch files to combine...")
    
    for batch_file in tqdm(batch_files, desc="Combining batches"):
        try:
            batch_path = os.path.join('all_batches_complete', batch_file)
            batch_df = pd.read_excel(batch_path)
            all_batches.append(batch_df)
        except Exception as e:
            print(f"Warning: Could not read {batch_file}: {e}")
    
    if all_batches:
        # Combine all batches
        master_df = pd.concat(all_batches, ignore_index=True)
        
        # Save master file
        master_filename = 'ZULU_LEXICON_COMPLETE.xlsx'
        master_df.to_excel(master_filename, index=False)
        
        print(f" Created master file: {master_filename}")
        print(f"   Total words: {len(master_df)}")
        print(f"   Columns: {list(master_df.columns)}")
        
        # Show sample
        print("\nSample of complete lexicon:")
        display(master_df[['CILUBA', 'French', 'Zulu']].head(10))
        
        return master_df
    else:
        print(" No batch files found to combine")
        return None

# Combine batches if we have any
if completed_batches > 0:
    complete_lexicon = combine_all_batches()
else:
    print("No batches were processed yet")
```
